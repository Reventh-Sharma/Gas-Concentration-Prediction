```python
import numpy as np
import pandas as pd
import math
from statistics import mean
```

```python
dataset = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Ethylene_CO1.csv')
```

```python
X = dataset.iloc[:, 4:].values
```

```python
X1=np.zeros((420*1900,16))
for i in range(420):
    X1[i*1900:(i+1)*1900,:]=X[i*10000+8000:(i+1)*10000-100,:]
```

```python
for z in range(int(np.size(X1,0)/1900)):
    print('\n',z)
    for i in range(1900):
        if i>249:
            for j in range(np.size(X1,1)):                         
                if X1[z*1900+i,j]>(X1[z*1900+i-1,j]+4*math.sqrt(mean(abs(X1[z*1900+i-250:z*1900+i,j] - X1[z*1900+i-1,j])**2))):
                    X1[z*1900+i,j]=X1[z*1900+i-1,j]+4*math.sqrt(mean(abs(X1[z*1900+i-250:z*1900+i,j] - X1[z*1900+i-1,j])**2))                                                
        print('=>',end='')
```

```python
for i in range(np.size(X1,0)):
    if i>79819:
        if X1[i,1]>350:
            X1[i,1]=mean(X1[i-250:i,1])
```

```python
df=pd.DataFrame(X1,index=np.arange(0,np.size(X1,0)),columns=np.arange(0,np.size(X1,1)))
```

```python
df.to_csv("/content/drive/My Drive/Colab Notebooks/unanomalousfinal.csv",index=False,header=False,index_label=False)
```
